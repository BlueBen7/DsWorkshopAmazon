## Bias Correction

### Helpful Tag Bias Correction

```python
### INFORMATION ###

# It is important to remove bias from the train set,
# i.e. that it will have equal number of reviews tagged as 'helpful'
# to the number of reviews tagged as 'unhelpful'. Otherwise,
# the classes won't be balanced and it can cause overfitting.
# Therefore, it's highly recommended to set 'correct_bias' to 'True'.

#@markdown ###Bias Correction
correct_bias = True #@param {type:"boolean"}

# Here we also choose the size of test set and it's percentage of the
# train and test samples altogether.
# It is recommended to choose 'test_set_size' between 15000 - 20000.

#@markdown ###Train-Test Split 
test_set_size = 15000  #@param {type:"slider", min: 1000, max: 35000, step:1000}
test_set_percentage = 0.3  #@param {type:"slider", min: 0.25, max: 0.45, step:0.025}
override_test_set = False       #@param {type: "boolean"}

### END OF INFORMATION ###
```

```python
### INFORMATION ###

# This cell prints the distribution of our classes, among the train and tests sets.

### END OF INFORMATION ###

# Add columns to mark reviews which are part of the test/validation test  
if 'isTestSet' in reviews.columns:
  reviews = reviews.drop(['isTestSet'], axis=1)
reviews.insert(0, 'isTestSet', False)

train_set_plain = reviews[(reviews['review_date'] > train_min_date) & (reviews['review_date'] < train_max_date)]

# Test set should be constant (only created once in the project lifetime)
if (override_test_set) or (os.path.exists(root + TEST_SET_PICKLE + "/" + pick_category) == False):
    print('Generating test set for the first time in the ds workflow lifetime!')
    test_set_plain = reviews[reviews['review_date'] > test_min_date] # test set + validation
     # mark examples used in test set
    if 'isTestSet' in reviews.columns:
      reviews = reviews.drop(['isTestSet'], axis=1)
    reviews.insert(0, 'isTestSet', True)
    pickleWrite(root, TEST_SET_PICKLE, test_set_plain, pick_category)
else:
    print('Loading existing test set for category %s' % pick_category)
    test_set_plain = pickleRead(root, TEST_SET_PICKLE, pick_category)
    
# Print train and test set information
train_info = get_set_info(train_set_plain)
test_info = get_set_info(test_set_plain)
print_set_info('Train', train_info)
print_set_info('Test', test_info)

###

num_test_no_neutral = test_info['helpful'] + test_info['unhelpful']
required_train_size = int(test_set_size * ((1 - test_set_percentage)/test_set_percentage))
train_size_after_correction = 2*min(train_info['helpful'],train_info['unhelpful'])

test_too_large_msg = '''There aren't enough reviews that can be used as a test set.
  The reason you see this error is because you requested a test size of '''+str(test_set_size)+''' reviews,
  but only '''+str(num_test_no_neutral)+''' reviews are available.\n
  In order to solve this error, you should increase the number of reviews available for testing by doing one of the following:
  1. Decrease 'test_set_size'.
  2. Go back to date selection (a few cells beore) and choose a bigger time range for the test set (earlier minimal date).
  3. Go back to the selection of the category and find a different category with more product reviews (such as DVD_Video).'''

cant_correct_bias_msg = '''Can\'t correct bias and keep test:train ratio.\n
    This error is generated on purpose (don't panic).
    The reason you see this error is because you chose to correct the bias of the train set,
    meaning there will be equal number of helpful and unhelpful reviews in the train set.
    The required train size is '''+str(required_train_size)+''' but there are only '''+str(train_size_after_correction)+''' reviews in the train set after bias correction.\n
    In order to solve this error, you should increase the number of reviews available for training by doing one of the following:
    1. Decrease 'test_set_percentage' or increase 'test_set_size'.
    2. Go back to date selection (a few cells beore) and choose a bigger time range for the train set.
    3. Go back to the selection of the category and find a different category with more product reviews (such as Music).'''

if (test_set_size>num_test_no_neutral):
  assert False, test_too_large_msg
  
if correct_bias:
  train_size_after_correction = 2*min(train_info['helpful'],train_info['unhelpful'])
  if (required_train_size>train_size_after_correction):
    assert False, cant_correct_bias_msg
```

```python
### INFORMATION ###

# This cell checks there are no untagged reviews in the train and test sets.
# It prints the final sizes of the train and test sets.

### END OF INFORMATION ###

print('Removing untagged reviews from the train and test sets.')
train_set_new = train_set_plain[train_set_plain['helpful'] != 0]
test_set_new = test_set_plain[test_set_plain['helpful'] != 0]

test_set_new = test_set_new.sample(test_set_size)

if correct_bias:
  #Guaranteed to have enough reviews (see assertion in previous cell).
  df1 = train_set_new[train_set_new['helpful'] == 1].sample(n=required_train_size//2)
  df2 = train_set_new[train_set_new['helpful'] == -1].sample(n=required_train_size//2)
  train_set_new = pd.concat([df1,df2])
else:
  train_set_new = train_set_new.sample(n=required_train_size)

# Mark the role of each data row in 'reviews' dataset.
reviews['role'] = 'UNUSED'
reviews.loc[train_set_new.index,'role'] = 'TRAIN'
reviews.loc[test_set_new.index,'role'] = 'TEST'
  
print("[TEST SET] model input (biased reviews):", test_set_new.shape[0])
print("[TRAIN SET] model input (unbiased reviews):", train_set_new.shape[0])
print("Total data size = unbiased train set reviews + biased test set reviews:", test_set_new.shape[0]+train_set_new.shape[0])
```

## Extract user data

```python
### INFORMATION ###

# This cell adds new features to the data called 'user data'.
# Its mainly statistics about the comments that were written by each reviewer.

### END OF INFORMATION ###

def get_user_data(df):
  #Keep this block of code under a function.
  commentCount=pd.DataFrame(df.groupby('customer_id').size())
  commentCount.columns=['commentCount']
  columns_before_user_data = list(df.columns.values)
  df = df.set_index('customer_id').join(commentCount)
  df = df.reset_index(drop=False) #drop=False to keep customer_id column.
  df['commentCount'] = df['commentCount'] - 1 #comment count should not include current comment
  helpfulRatingPerUser = pd.DataFrame(df.groupby('customer_id').sum())
  helpfulRatingPerUser = helpfulRatingPerUser.drop(columns=['isTestSet','product_parent','star_rating','commentCount'])
  helpfulRatingPerUser.columns=['total_helpful_votes_for_user','total_votes_for_user','userHelpfulness']
  df = df.join(helpfulRatingPerUser, on='customer_id')
  df['userHelpfulness'] = df['userHelpfulness'] - df['helpful']
  df['total_helpful_votes_for_user'] = df['total_helpful_votes_for_user'] - df['helpful_votes']
  df['total_votes_for_user'] = df['total_votes_for_user'] - df['total_votes']
  df['reviewer_avg_rating'] = df['total_helpful_votes_for_user']/df['total_votes_for_user']
  df['reviewer_avg_rating'] = df['reviewer_avg_rating'].fillna(0)
  columns_after_user_data = list(df.columns.values)
  diff_columns = list(set(columns_after_user_data) - set(columns_before_user_data))
  print("Added the following columns: ", ', '.join(diff_columns))
  return df

reviews_with_user_data = get_user_data(reviews)
```

## Feature selection using a wrapper method

```python
### INFORMATION ###

# This cell chooses the best 10 numeric features from all the generated features.
# It uses Sequential Backward Selection (SBS), which is a wrapper method.

# The numeric features are more useful when they have gaussian-like distribution.
# Therefore, we apply power transform before the wrapper method (this is part of the pipeline),
# and the old and new distributions of the features are displayed in a graph.

### END OF INFORMATION ###

def plot_hists(dataset, nrows, ncols, title=""):
  original_cols = list(dataset.columns)
  plt.figure(figsize=(12, 16))
  for i in range(len(original_cols)):
    colname = original_cols[i]
    plt.subplot(nrows, ncols, i+1)
    plt.title(colname)
    dataset[colname].plot.hist(bins=200)
  plt.suptitle(title)
  plt.tight_layout()
  plt.subplots_adjust(top=0.95)
  plt.show()
  
def transform_data(dataset):
  original_cols = list(dataset.columns)
  power_trans = PowerTransformer(copy=True, method='yeo-johnson', standardize=True) 
  transf_data = power_trans.fit_transform(dataset)
  df = pd.DataFrame(data=transf_data, columns=original_cols)
  return df

all_numeric_features = [
 'flesch_reading_ease',
 'smog_index',
 'flesch_kincaid_grade',
 'coleman_liau_index',
 'automated_readability_index',
 'dale_chall_readability_score',
 'difficult_words',
 'linsear_write_formula',
 'gunning_fog',
 'commentCount',
 'userHelpfulness',
 'body_length',
 'body_num_words',
 'num_spelling_mistakes',
 'total_helpful_votes_for_user',
 'total_votes_for_user',
 'reviewer_avg_rating',
 'subjectivity_avg',
 'subjectivity_std',
 'polarity_avg',
 'polarity_std',
]

numeric_X_train = train_set_new[all_numeric_features].astype('f').reset_index(drop=True)
numeric_Y_train = train_set_new['helpful'].reset_index(drop=True)

# Transform data.
plot_hists(numeric_X_train, 7, 3, "Original Distributions") 
numeric_X_train = transform_data(numeric_X_train)
plot_hists(numeric_X_train, 7, 3, "Modified Distributions")   

sfs = SequentialFeatureSelector(RandomForestClassifier(n_estimators=20, max_depth=4),
                                k_features=10, forward=False, verbose=2, scoring='f1', cv=3)
sfs = sfs.fit(numeric_X_train.values, numeric_Y_train)
best_k_numeric = [all_numeric_features[feature_index] for feature_index in sfs.k_feature_idx_]

print('\nSequential Backward Selection Result:')
print(best_k_numeric)
print('CV Score:', sfs.k_score_)
```

```python
### INFORMATION ###

# This cell splits the train and test sets' features and labels into variables.

### END OF INFORMATION ###

input_features = best_k_numeric + ['review_headline', 'review_body']
label = 'helpful'

# Get the final train and test sets.
final_train_set = data[data['role']=='TRAIN']
final_test_set = data[data['role']=='TEST']

# Make sure there is no indication for test or train in data.
if 'role' in final_train_set:
  final_train_set = final_train_set.drop(['role'], axis=1)

if 'role' in final_test_set:
  final_test_set = final_test_set.drop(['role'], axis=1)

# Split the sets to features (X) and label (Y).
X_train = final_train_set[input_features]
Y_train = final_train_set[label]
X_test = final_test_set[input_features]
Y_test = final_test_set[label]
```

## Hyperparameter Tuning Using Grid Search Cross Validation

```python
### INFORMATION ###

# The cells in this section perform tuning of different hyperparameters of the models,
# such as optimizing the parameters of the different classifiers.

# This cell finds the best parameters for the numerical pipeline and random forest classifier.

### END OF INFORMATION ###

numeric_pipeline_no_clf = Pipeline([
  ('select', FunctionTransformer(lambda x: x[best_k_numeric].astype('f'), validate=False)),
  ('pt', PowerTransformer(copy=True, method='yeo-johnson', standardize=True)),
])

X_train_clf_ready = numeric_pipeline_no_clf.fit_transform(X_train)

param_grid = { 
    'n_estimators': [150, 175, 200],
    'max_depth' : [4, 5],
} 

param_grid_obj = ParameterGrid(param_grid)
all_param_combi = [comb for comb in param_grid_obj]
num_combs = len(all_param_combi)
print('Have', num_combs, 'parameter combinations.')

random_forest = RandomForestClassifier(n_estimators=100, max_depth=2)

best_score = 0
best_grid = {}

num_cv = 3

for i in range(num_combs):
  comb = all_param_combi[i]
  random_forest = random_forest.set_params(**comb)
  scores = []
  for j in range(num_cv):
    random_forest.fit(X_train_clf_ready, Y_train)

    y = random_forest.predict(X_train_clf_ready)
    score = metrics.f1_score(Y_train, y)
    scores.append(score)
  avg_run_score = np.mean(scores)
  
  # Keep track of the best result.
  if avg_run_score > best_score:
    best_score = score
    best_grid = comb
  print("Finished", i+1, "out of", num_combs, ". Best score:", best_score)
  print("Best combination:", best_grid)

# Use best parameter combination.
random_forest = random_forest.set_params(**best_grid) 
```

```python
### INFORMATION ###

# This cell finds the best parameters for the textual pipeline and logistic regression classifier.

### END OF INFORMATION ###

# Subclass of SelectKBest
class SelectAtMostKBest(SelectKBest):
  def _check_params(self, X, y):
      if not (self.k == "all" or 0 <= self.k <= X.shape[1]):
          # Set k to "all" (skip feature selection)
          # if less than k features are available.
          self.k = "all"

#Tune textual model

tfidf_head = TfidfVectorizer(sublinear_tf=True, max_df=0.5, ngram_range=(1, 2),
                                    tokenizer=best_tokenizer, analyzer='word')
tfidf_body = TfidfVectorizer(sublinear_tf=True, max_df=0.5, ngram_range=(1, 2),
                                    tokenizer=best_tokenizer, analyzer='word') #min-df=5?

head_pipeline = Pipeline([
    ('select', FunctionTransformer(lambda x: x['review_headline'].astype('U'), validate=False)),
    ('tfidf', tfidf_head),
    ('kbest', SelectAtMostKBest(chi2, k=1000)),
])

body_pipeline = Pipeline([
    ('select', FunctionTransformer(lambda x: x['review_body'].astype('U'), validate=False)),
    ('tfidf', tfidf_body),
    ('kbest', SelectAtMostKBest(chi2, k=1000)),
])

textual_pipeline_no_clf = FeatureUnion(
  transformer_list=[
      ('head', head_pipeline),
      ('body', body_pipeline),
  ]
)

X_train_clf_ready = textual_pipeline_no_clf.fit_transform(X_train, Y_train)
```

```python
### INFORMATION ###

# Continue.

### END OF INFORMATION ###

param_grid = { 
    'C': [1] + [x*5 for x in range(1,20)],
} 

param_grid_obj = ParameterGrid(param_grid)
all_param_combi = [comb for comb in param_grid_obj]
num_combs = len(all_param_combi)
print('Have', num_combs, 'parameter combinations.')

logistic_reg = LogisticRegression(solver='liblinear')

best_score = 0
best_grid = {}

num_cv = 3

for i in range(num_combs):
  comb = all_param_combi[i]
  logistic_reg = logistic_reg.set_params(**comb)
  scores = []
  for j in range(num_cv):
    logistic_reg.fit(X_train_clf_ready, Y_train)

    y = logistic_reg.predict(X_train_clf_ready)
    score = metrics.f1_score(Y_train, y)
    scores.append(score)
  avg_run_score = np.mean(scores)
  
  # Keep track of the best result.
  if avg_run_score > best_score:
    best_score = score
    best_grid = comb
  print("Finished", i+1, "out of", num_combs, ". Best score:", best_score)
  print("Best combination:", best_grid)

# Use best parameter combination.
logistic_reg = logistic_reg.set_params(**best_grid) 
```

## Pipeline Definition

```python
### INFORMATION ###

# This cell defines the pipeline of the model.

### END OF INFORMATION ###

### Textual Model ###
feature_union = FeatureUnion(
  transformer_list=[
      ('head', Pipeline([
          ('select', FunctionTransformer(lambda x: x['review_headline'].astype('U'), validate=False)),
          ('tfidf', tfidf_head),
          ('kbest', SelectAtMostKBest(chi2, k=1000000)),
      ])), 

      ('body', Pipeline([
          ('select', FunctionTransformer(lambda x: x['review_body'].astype('U'), validate=False)),
          ('tfidf', tfidf_body),
          ('kbest', SelectAtMostKBest(chi2, k=1000000)),
      ])),
  ]
)

pipeline_textual = Pipeline([
  ('union', feature_union),  #tf-idf is inside.
  ('kbest', SelectKBest(chi2, k='all')),
  ('clf', logistic_reg),
])

### Numerical Model ###

pipeline_numerical = Pipeline([
  ('select', FunctionTransformer(lambda x: x[best_k_numeric].astype('f'), validate=False)),
  ('names', GetNames()),
  ('pt', PowerTransformer(copy=True, method='yeo-johnson', standardize=True)),
  ('kbest', SelectKBest(f_classif, k='all')),
  ('clf', random_forest),
])

### Soft Voting Classifier ###

voting_clf = VotingClassifier(
	estimators = [
      ('pipe_num', pipeline_numerical),
      ('pipe_txt', pipeline_textual),
  ],
	voting = 'soft')
```

## Model Fitting and Testing

```python
### INFORMATION ###

# The model is already configured and its parameters are tuned.
# This cell trains and tests the model.
# It is recommended to use cross validation (which takes a bit more time).
# This cell takes about 5 minutes to complete.

use_cross_validation = True

### END OF INFORMATION ###

if use_cross_validation:
  # Train the model using k-Fold Cross-Validation,
  # which we implement using GridSearchCV, given an empty dictionary.
  model = GridSearchCV(voting_clf, {}, iid=False, cv=3,
                       return_train_score=False, scoring='f1')
else:
  model = voting_clf

  
### Train Model ###

model.fit(X_train, Y_train)

### Test Model ###

y = model.predict(X_test)
print_scores(Y_test, y)
```
